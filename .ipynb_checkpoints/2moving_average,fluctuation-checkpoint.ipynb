```python
import numpy as np
import pandas as pd
from pathlib import Path

def preprocess_timestamps(df):
    """
    Preprocess timestamps by removing zero values and reindexing time stamps
    to group into sets of 38 readings per day while preserving original time values.
    
    Args:
        df: DataFrame with 'time' and 'value' columns
    
    Returns:
        DataFrame with additional 'original_time', 'day_number' and 'time_index' columns
    """
    # Store original time values in a new column
    df['original_time'] = df['time']
    
    # Remove rows where value is zero or negative
    df = df[df['value'] > 0].reset_index(drop=True)
    
    # Reset time column to start from 1
    df['time'] = df.index + 1
    
    # Calculate day number (1-based) and time index (1-38)
    # Each complete day has 38 readings
    df['day_number'] = (df.index // 38) + 1
    df['time_index'] = df.index % 38 + 1
    
    return df

def calculate_moving_average_by_timestamp(df, window_days=21):
    """
    Calculate moving average preserving intraday patterns.
    For each time index (1-38), calculates average using the same time index
    across different days within the window.
    
    Args:
        df: DataFrame with 'time', 'value', 'day_number', and 'time_index' columns
        window_days: Number of days for moving average window (default: 21)
    """
    # Initialize array for moving averages
    ma_values = np.full(len(df), np.nan)
    
    # Calculate half window size
    half_window = window_days // 2
    
    # Process each time index (1-38) separately
    for time_idx in range(1, 39):  # 38 readings per day
        # Get all values for this specific time of day
        mask = df['time_index'] == time_idx
        time_specific_data = df[mask].copy()
        
        if len(time_specific_data) > 0:
            # Sort by day number to ensure chronological order
            time_specific_data = time_specific_data.sort_values('day_number')
            
            # Get positions in original dataframe for this time index
            original_positions = df.index[mask].values
            
            # Calculate moving average for each day at this time index
            for i in range(len(time_specific_data)):
                current_day = time_specific_data.iloc[i]['day_number']
                
                # Skip first and last 10 days (not enough data for 21-day window)
                if current_day <= half_window or current_day > (df['day_number'].max() - half_window):
                    ma_values[original_positions[i]] = np.nan
                    continue
                
                # Calculate window boundaries
                start_idx = max(0, i - half_window)
                end_idx = min(len(time_specific_data), i + half_window + 1)
                
                # Get values within window (same time of day across different days)
                window_values = time_specific_data['value'].iloc[start_idx:end_idx]
                
                # Only calculate average if we have exactly window_days points
                if len(window_values) == window_days:
                    ma_values[original_positions[i]] = window_values.mean()
                else:
                    ma_values[original_positions[i]] = np.nan
    
    return pd.Series(ma_values, index=df.index)



def process_time_series_data(input_path):
    """
    Process time series data file(s) for stock market analysis.
    """
    def process_single_file(file_path, output_folder):
        # Load the data
        df = pd.read_csv(file_path, delim_whitespace=True, header=None, names=['time', 'value'])
        
        print(f"\nProcessing {file_path}")
        print(f"Initial rows: {len(df)}")
        print(f"Zero or negative values: {len(df[df['value'] <= 0])}")
        
        # Preprocess data (remove zeros and reindex)
        df = preprocess_timestamps(df)
        print(f"Rows after removing zeros: {len(df)}")
        
        # Verify time structure
        print(f"\nTime structure:")
        print(f"Number of days: {df['day_number'].max()}")
        print(f"Readings per day: {len(df['time_index'].unique())}")
        
        # Calculate moving average
        df['moving_average'] = calculate_moving_average_by_timestamp(df)
        
        # Calculate fluctuations
        df['fluctuation'] = df['value'] - df['moving_average']
        
        # Remove rows with NaN in moving_average
        df_cleaned = df.dropna(subset=['moving_average'])
        print(f"Final rows after cleaning: {len(df_cleaned)}")
        
        # Print sample data for verification
        print("\nSample data for first time index (first 5 days):")
        sample = df_cleaned[df_cleaned['time_index'] == 1].head()
        print(sample[['day_number', 'time_index', 'value', 'moving_average']])
        
        # Save processed data
        output_path = output_folder / f"{Path(file_path).stem}.csv"
        df_cleaned.to_csv(output_path, index=False)
        return str(output_path)

    # Handle input path
    path = Path(input_path)
    output_folder = path / "processed_data"
    output_folder.mkdir(parents=True, exist_ok=True)
    
    processed_files = []
    
    if path.is_file():
        if path.suffix == '.dat':
            processed_files.append(process_single_file(str(path), output_folder))
    elif path.is_dir():
        for file_path in path.glob("*.dat"):
            processed_files.append(process_single_file(str(file_path), output_folder))
        if not processed_files:
            raise ValueError(f"No .dat files found in directory {path}")
    else:
        raise ValueError(f"Input path {path} does not exist")
    
    return processed_files

# Example usage:
processed_files = process_time_series_data('Dataset')
```

# Plotting


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

def plot_distribution_analysis(distribution, method, base_path='Dataset/processed_data', save_plot=False):
    """
    Create time series plots for a given distribution and method, showing both theta and phi parameters.
    
    Args:
        distribution (str): Name of the distribution (gamma, inverseGamma, logNormal, weibull)
        method (str): Estimation method (MLE, LeastSquare)
        base_path (str): Base path to the processed CSV files
        save_plot (bool): Whether to save the plot to disk (default: False)
    """
    # Set style for better visualization
    sns.set_theme()
    
    # Create figure with 6 subplots (3 for theta, 3 for phi)
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    fig.suptitle(f'{distribution} Distribution - {method} Method', fontsize=16, y=0.95)
    
    # Parameters to plot
    params = ['theta', 'phi']
    metrics = ['value', 'moving_average', 'fluctuation']
    
    # Load and plot data for each parameter
    for param_idx, param in enumerate(params):
        # Construct file path
        file_name = f"{distribution}_{method}_{param}.csv"
        file_path = Path(base_path) / file_name
        
        # Load data
        try:
            df = pd.read_csv(file_path)
        except FileNotFoundError:
            print(f"File not found: {file_path}")
            continue
            
        # Plot each metric
        for metric_idx, metric in enumerate(metrics):
            ax = axes[param_idx, metric_idx]
            
            # Plot the data
            ax.plot(df['time'], df[metric], linewidth=1)
            
            # Set labels and title
            ax.set_xlabel('Time')
            ax.set_ylabel(metric.replace('_', ' ').title())
            ax.set_title(f'{param.title()} - {metric.replace("_", " ").title()}')
            
            # Rotate x-axis labels for better readability
            ax.tick_params(axis='x', rotation=45)
            
            # Add grid
            ax.grid(True, alpha=0.3)
    
    # Adjust layout to prevent overlap
    plt.tight_layout()
    
    # Save plot if requested
    if save_plot:
        output_dir = Path('Results/Plots')
        output_dir.mkdir(parents=True, exist_ok=True)
        fig.savefig(output_dir / f'{distribution}_{method}_analysis.png', dpi=300, bbox_inches='tight')
    
    plt.show()
    return fig

def plot_all_distributions(save_plot=False):
    """
    Create plots for all distributions and methods.
    
    Args:
        save_plot (bool): Whether to save the plots to disk (default: False)
    """
    distributions = ['gamma', 'InvG', 'LogN', 'Weib']
    methods = ['MLE', 'LeastSquare']
    
    # Create plots for each combination
    for dist in distributions:
        for method in methods:
            fig = plot_distribution_analysis(dist, method, save_plot=save_plot)
            plt.close(fig)

# Example usage
# For displaying plots only:
plot_all_distributions(save_plot=False)

# For saving plots:
# plot_all_distributions(save_plot=True)
```

```python
import numpy as np
import pandas as pd
from pathlib import Path

def preprocess_timestamps(df):
    """
    Preprocess timestamps by removing zero values and reindexing time stamps
    to group into sets of 38 readings per day.
    
    Args:
        df: DataFrame with 'time' and 'value' columns
    
    Returns:
        DataFrame with additional 'day_number' and 'time_index' columns
    """
    # Remove rows where value is zero or negative
    df = df[df['value'] > 0].reset_index(drop=True)
    
    # Calculate day number (1-based) and time index (1-38)
    # Each complete day has 38 readings
    df['day_number'] = (df.index // 38) + 1
    df['time_index'] = df.index % 38 + 1
    
    return df

def calculate_moving_average_by_timestamp(df, window_days=21):
    """
    Calculate moving average preserving intraday patterns.
    For each time index (1-38), calculates average using the same time index
    across different days within the window.
    
    Args:
        df: DataFrame with 'time', 'value', 'day_number', and 'time_index' columns
        window_days: Number of days for moving average window (default: 21)
    """
    # Initialize array for moving averages
    ma_values = np.full(len(df), np.nan)
    
    # Calculate half window size
    half_window = window_days // 2
    
    # Process each time index (1-38) separately
    for time_idx in range(1, 39):  # 38 readings per day
        # Get all values for this specific time of day
        mask = df['time_index'] == time_idx
        time_specific_data = df[mask].copy()
        
        if len(time_specific_data) > 0:
            # Sort by day number to ensure chronological order
            time_specific_data = time_specific_data.sort_values('day_number')
            
            # Get positions in original dataframe for this time index
            original_positions = df.index[mask].values
            
            # Calculate moving average for each day at this time index
            for i in range(len(time_specific_data)):
                # Calculate window boundaries
                start_idx = max(0, i - half_window)
                end_idx = min(len(time_specific_data), i + half_window + 1)
                
                # Get values within window (same time of day across different days)
                window_values = time_specific_data['value'].iloc[start_idx:end_idx]
                
                # Only calculate average if we have enough data points
                if len(window_values) >= half_window:
                    ma_values[original_positions[i]] = window_values.mean()
    
    return pd.Series(ma_values, index=df.index)

def verify_calculations(df_cleaned, specific_day=10, specific_time=1, window_size=21):
    """
    Verify the moving average calculations for specific time indices.
    
    Args:
        df_cleaned: Processed DataFrame with moving averages
        specific_day: Day number to verify
        specific_time: Time index to verify
        window_size: Size of moving average window
    """
    print("\nDetailed Verification of Moving Average Calculations:")
    print("-" * 50)
    
    # Check specific time index across days
    time_idx_to_check = specific_time
    df_check = df_cleaned[df_cleaned['time_index'] == time_idx_to_check].sort_values('day_number')
    print(f"\nChecking all values for time index {time_idx_to_check} across first 10 days:")
    print(df_check[['day_number', 'time_index', 'value', 'moving_average']].head(10))
    
    # Verify moving average calculation for specific day and time
    half_window = window_size // 2
    df_window = df_cleaned[
        (df_cleaned['time_index'] == specific_time) & 
        (df_cleaned['day_number'].between(specific_day - half_window, specific_day + half_window))
    ]
    
    print(f"\nVerifying moving average calculation for day {specific_day}, time index {specific_time}:")
    print(f"Window size: {window_size} days")
    print(f"Number of values in window: {len(df_window)}")
    print(f"Manual average: {df_window['value'].mean():.2f}")
    
    calculated_ma = df_cleaned[
        (df_cleaned['day_number'] == specific_day) & 
        (df_cleaned['time_index'] == specific_time)
    ]['moving_average'].iloc[0]
    print(f"Calculated MA: {calculated_ma:.2f}")
    
    print("\nValues used in window:")
    print(df_window[['day_number', 'value']].to_string())

def process_time_series_data(input_path):
    """
    Process time series data file(s) for stock market analysis.
    """
    def process_single_file(file_path, output_folder):
        # Load the data
        df = pd.read_csv(file_path, delim_whitespace=True, header=None, names=['time', 'value'])
        
        print(f"\nProcessing {file_path}")
        print(f"Initial rows: {len(df)}")
        print(f"Zero or negative values: {len(df[df['value'] <= 0])}")
        
        # Preprocess data (remove zeros and reindex)
        df = preprocess_timestamps(df)
        print(f"Rows after removing zeros: {len(df)}")
        
        # Verify time structure
        print(f"\nTime structure:")
        print(f"Number of days: {df['day_number'].max()}")
        print(f"Readings per day: {len(df['time_index'].unique())}")
        
        # Calculate moving average
        df['moving_average'] = calculate_moving_average_by_timestamp(df)
        
        # Calculate fluctuations
        df['fluctuation'] = df['value'] - df['moving_average']
        
        # Remove rows with NaN in moving_average
        df_cleaned = df.dropna(subset=['moving_average'])
        print(f"Final rows after cleaning: {len(df_cleaned)}")
        
        # Print sample data for verification
        print("\nSample data for first time index (first 5 days):")
        sample = df_cleaned[df_cleaned['time_index'] == 1].head()
        print(sample[['day_number', 'time_index', 'value', 'moving_average']])
        
        # Run detailed verification
        verify_calculations(df_cleaned)
        
        # Save processed data
        output_path = output_folder / f"{Path(file_path).stem}.csv"
        df_cleaned.to_csv(output_path, index=False)
        return str(output_path)

    # Handle input path
    path = Path(input_path)
    output_folder = path / "processed_data"
    output_folder.mkdir(parents=True, exist_ok=True)
    
    processed_files = []
    
    if path.is_file():
        if path.suffix == '.dat':
            processed_files.append(process_single_file(str(path), output_folder))
    elif path.is_dir():
        for file_path in path.glob("*.dat"):
            processed_files.append(process_single_file(str(file_path), output_folder))
        if not processed_files:
            raise ValueError(f"No .dat files found in directory {path}")
    else:
        raise ValueError(f"Input path {path} does not exist")
    
    return processed_files

# Example usage:
processed_files = process_time_series_data('Dataset')
```
